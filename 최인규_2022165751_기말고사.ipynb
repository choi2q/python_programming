```python
import requests
import pandas as pd

def download_data(url):
    response = requests.get(url)
    data = response.json()
    df = pd.DataFrame(data["response"]["docs"])
    df = df["doc"].apply(pd.Series)
    return df

# 2017년 5월 데이터 다운로드
url_2017 = 'http://data4library.kr/api/loanItemSrch?libCode=111344&format=json&startDt=2017-05-01&endDt=2017-05-31&dtl_region=11040&authKey=c01ec15e4574f74ee45cba2601bad15b82971e606e3b0740977ee4b363ce2fe2'
df_2017 = download_data(url_2017)

# 2023년 5월 데이터 다운로드
url_2023 = 'http://data4library.kr/api/loanItemSrch?libCode=111344&format=json&startDt=2023-05-01&endDt=2023-05-31&dtl_region=11040&authKey=c01ec15e4574f74ee45cba2601bad15b82971e606e3b0740977ee4b363ce2fe2'
df_2023 = download_data(url_2023)

# Convert 'loan_count' column to integers
df_2017['loan_count'] = pd.to_numeric(df_2017['loan_count'])
df_2023['loan_count'] = pd.to_numeric(df_2023['loan_count'])

# 대출 횟수가 가장 많은 상위 20권 추출 (2017년과 2023년 모두에서)
top_20_2017 = df_2017.nlargest(20, "loan_count")
top_20_2023 = df_2023.nlargest(20, "loan_count")

# 상위 20권 도서 리스트 출력
print("2017년 5월 상위 20권 도서:")
print(top_20_2017[["bookname", "authors", "publisher", "publication_year", "isbn13", "loan_count"]].to_string(index=False))
print()

print("2023년 5월 상위 20권 도서:")
print(top_20_2023[["bookname", "authors", "publisher", "publication_year", "isbn13", "loan_count"]].to_string(index=False))
print()

# 빈도수 비교
booknames_2017 = top_20_2017["bookname"].str.strip().tolist()
booknames_2023 = top_20_2023["bookname"].str.strip().tolist()

bookname_freq_2017 = {bookname: booknames_2017.count(bookname) for bookname in booknames_2017}
bookname_freq_2023 = {bookname: booknames_2023.count(bookname) for bookname in booknames_2023}

print("도서명 빈도수 비교:")
for bookname in booknames_2017:
    freq_2017 = bookname_freq_2017[bookname]
    freq_2023 = bookname_freq_2023.get(bookname, 0)
    print(f"{bookname}: {freq_2017} (2017년), {freq_2023} (2023년)")

```

```python
import requests
import pandas as pd
from tabulate import tabulate

def download_data(url):
    response = requests.get(url)
    data = response.json()
    df = pd.DataFrame(data["response"]["docs"])
    df = df["doc"].apply(pd.Series)
    return df

# 2017년 5월 데이터 다운로드
url_2017 = 'http://data4library.kr/api/loanItemSrch?libCode=111344&format=json&startDt=2017-05-01&endDt=2017-05-31&dtl_region=11040&authKey=c01ec15e4574f74ee45cba2601bad15b82971e606e3b0740977ee4b363ce2fe2'
df_2017 = download_data(url_2017)

# 2023년 5월 데이터 다운로드
url_2023 = 'http://data4library.kr/api/loanItemSrch?libCode=111344&format=json&startDt=2023-05-01&endDt=2023-05-31&dtl_region=11040&authKey=c01ec15e4574f74ee45cba2601bad15b82971e606e3b0740977ee4b363ce2fe2'
df_2023 = download_data(url_2023)

# Convert 'loan_count' column to numeric type
df_2017['loan_count'] = pd.to_numeric(df_2017['loan_count'])
df_2023['loan_count'] = pd.to_numeric(df_2023['loan_count'])

# 대출 횟수가 가장 많은 상위 20권 추출 (2017년과 2023년 모두에서)
top_20_2017 = df_2017.nlargest(20, "loan_count")
top_20_2023 = df_2023.nlargest(20, "loan_count")

# 도서명을 기준으로 빈도수 계산 (2017년과 2023년 모두에서)
bookname_freq_2017 = top_20_2017["bookname"].value_counts().to_dict()
bookname_freq_2023 = top_20_2023["bookname"].value_counts().to_dict()

# 테이블 생성을 위한 데이터 준비
table_rows = []

for bookname in set(list(bookname_freq_2017.keys()) + list(bookname_freq_2023.keys())):
    freq_2017 = bookname_freq_2017.get(bookname, 0)
    freq_2023 = bookname_freq_2023.get(bookname, 0)
    table_rows.append([bookname, freq_2017, freq_2023])

# Print the table using the tabulate function
table_headers = ["도서명", "빈도수(2017)", "빈도수(2023)"]
table = tabulate(table_rows, headers=table_headers, tablefmt="pipe")

print(table)

```

```python
import requests
import pandas as pd
from tabulate import tabulate

# 데이터 다운로드
def download_data(year, month):
    url = f"http://data4library.kr/api/loanItemSrch?libCode=111344&format=json&startDt={year}-{month:02d}-01&endDt={year}-{month:02d}-31&dtl_region=11040&authKey=c01ec15e4574f74ee45cba2601bad15b82971e606e3b0740977ee4b363ce2fe2"
    response = requests.get(url)
    data = response.json()
    df = pd.DataFrame(data["response"]["docs"])
    df = df["doc"].apply(pd.Series)
    return df

# 2017년 5월 데이터 다운로드
df_2017 = download_data(2017, 5)

# 2023년 5월 데이터 다운로드
df_2023 = download_data(2023, 5)

# 'loan_count' 열을 숫자형으로 변환
df_2017['loan_count'] = pd.to_numeric(df_2017['loan_count'])
df_2023['loan_count'] = pd.to_numeric(df_2023['loan_count'])

# 대출 횟수를 기준으로 내림차순으로 데이터 정렬하고, 각 연도별 상위 100개 도서 선택
top_100_2017 = df_2017.nlargest(100, "loan_count")
top_100_2023 = df_2023.nlargest(100, "loan_count")

# 상위 100개 목록에서 공통된 도서 이름 가져오기
common_books = set(top_100_2017['bookname']).intersection(top_100_2023['bookname'])

# 공통된 도서와 대출 횟수를 저장할 데이터프레임 생성
merged_df = pd.DataFrame({'Book': list(common_books)})

# 2017년과 2023년의 대출 횟수 병합
merged_df = merged_df.merge(top_100_2017[['bookname', 'loan_count']], left_on='Book', right_on='bookname', how='left')
merged_df = merged_df.merge(top_100_2023[['bookname', 'loan_count']], left_on='Book', right_on='bookname', how='left')

# 대출 횟수 열 이름 변경
merged_df = merged_df.rename(columns={'loan_count_x': 'Loan count in 2017', 'loan_count_y': 'Loan count in 2023'})

# '2017년 대출 횟수' 열을 기준으로 내림차순 정렬
merged_df = merged_df.sort_values('Loan count in 2017', ascending=False)

# 테이블에 표시할 열 선택
selected_columns = ['Book', 'Loan count in 2017', 'Loan count in 2023']

# tabulate 함수를 사용하여 테이블 출력
table = tabulate(merged_df[selected_columns], headers=selected_columns, tablefmt='pipe', showindex=False)

print(table)
```

```python
import matplotlib.pyplot as plt
import requests
import pandas as pd

plt.rc('font', family='NanumBarunGothic')

def download_data(url):
    response = requests.get(url)
    data = response.json()
    df = pd.DataFrame(data["response"]["docs"])
    df = df["doc"].apply(pd.Series)
    return df

# 2017년 5월 데이터 다운로드
url_2017 = 'http://data4library.kr/api/loanItemSrch?libCode=111344&format=json&startDt=2017-05-01&endDt=2017-05-31&dtl_region=11040&authKey=c01ec15e4574f74ee45cba2601bad15b82971e606e3b0740977ee4b363ce2fe2'
df_2017 = download_data(url_2017)

# 2023년 5월 데이터 다운로드
url_2023 = 'http://data4library.kr/api/loanItemSrch?libCode=111344&format=json&startDt=2023-05-01&endDt=2023-05-31&dtl_region=11040&authKey=c01ec15e4574f74ee45cba2601bad15b82971e606e3b0740977ee4b363ce2fe2'
df_2023 = download_data(url_2023)

# 'loan_count' 열을 숫자 형식으로 변환
df_2017['loan_count'] = pd.to_numeric(df_2017['loan_count'])
df_2023['loan_count'] = pd.to_numeric(df_2023['loan_count'])

# 'no' 열을 기준으로 데이터프레임 병합
merged_df = pd.merge(df_2017, df_2023, on='no', suffixes=('_2017', '_2023'))

# 'loan_count_2017' 열을 기준으로 데이터프레임 내림차순 정렬 후 상위 50권 선택
top_50_books_2017 = merged_df.sort_values('loan_count_2017', ascending=False).head(50)

# 2017년과 2023년 상위 50권의 대출 횟수 비교
top_50_books_2017['loan_count_2023'] = merged_df.loc[top_50_books_2017.index, 'loan_count_2023']

# 2017년과 2023년의 대출 횟수 차이 계산
top_50_books_2017['count_change'] = top_50_books_2017['loan_count_2023'] - top_50_books_2017['loan_count_2017']

# 대출 횟수가 증가한 도서와 감소한 도서로 분리
increased_books = top_50_books_2017[top_50_books_2017['count_change'] > 0]
decreased_books = top_50_books_2017[top_50_books_2017['count_change'] < 0]

# 각 도서의 대출 횟수 차이를 막대 그래프로 표시
plt.figure(figsize=(12, 6))
plt.bar(increased_books['bookname_2017'], increased_books['count_change'], color='green', label='증가')
plt.bar(decreased_books['bookname_2017'], decreased_books['count_change'], color='red', label='감소')
plt.xticks(rotation=90)
plt.xlabel('도서')
plt.ylabel('대출 횟수 변화')
plt.title('2017년과 2023년 5월의 상위 50권 도서 대출 횟수 비교')
plt.legend()
plt.rcParams['axes.unicode_minus'] = False  # 음수 기호(-) 표시 설정
plt.savefig('book_loan_comparison.png', dpi=300, bbox_inches='tight', pad_inches=0.5, transparent=True)
plt.show()
```

```python
  import matplotlib.pyplot as plt
import requests
import pandas as pd

plt.rc('font', family='NanumBarunGothic')

def download_data(url):
    response = requests.get(url)
    data = response.json()
    df = pd.DataFrame(data["response"]["docs"])
    df = df["doc"].apply(pd.Series)
    return df

# 2017년 5월 데이터 다운로드
url_2017 = 'http://data4library.kr/api/loanItemSrch?libCode=111344&format=json&startDt=2017-05-01&endDt=2017-05-31&dtl_region=11040&authKey=c01ec15e4574f74ee45cba2601bad15b82971e606e3b0740977ee4b363ce2fe2'
df_2017 = download_data(url_2017)

# 2023년 5월 데이터 다운로드
url_2023 = 'http://data4library.kr/api/loanItemSrch?libCode=111344&format=json&startDt=2023-05-01&endDt=2023-05-31&dtl_region=11040&authKey=c01ec15e4574f74ee45cba2601bad15b82971e606e3b0740977ee4b363ce2fe2'
df_2023 = download_data(url_2023)

# 'loan_count' 열을 숫자 형식으로 변환
df_2017['loan_count'] = pd.to_numeric(df_2017['loan_count'])
df_2023['loan_count'] = pd.to_numeric(df_2023['loan_count'])

# 출판사별 대출 횟수 계산
loan_count_2017_by_publisher = df_2017.groupby('publisher')['loan_count'].sum()
loan_count_2023_by_publisher = df_2023.groupby('publisher')['loan_count'].sum()

# 출판사별 대출 횟수 차이 계산
loan_count_change_by_publisher = loan_count_2023_by_publisher - loan_count_2017_by_publisher

# 대출 횟수가 증가한 출판사와 감소한 출판사로 분리
increased_publishers = loan_count_change_by_publisher[loan_count_change_by_publisher > 0]
decreased_publishers = loan_count_change_by_publisher[loan_count_change_by_publisher < 0]

# 출판사별 대출 횟수 차이를 막대 그래프로 표시
plt.figure(figsize=(12, 6))
plt.bar(increased_publishers.index, increased_publishers, color='green', label='증가')
plt.bar(decreased_publishers.index, decreased_publishers, color='red', label='감소')
plt.xticks(rotation=90)
plt.xlabel('출판사')
plt.ylabel('대출 횟수 변화')
plt.title('2017년과 2023년 5월의 출판사별 도서 대출 횟수 비교')
plt.legend()
plt.rcParams['axes.unicode_minus'] = False  # 음수 기호(-) 표시 설정
plt.savefig('book_loan_comparison.png', dpi=300, bbox_inches='tight', pad_inches=0.5, transparent=True)
plt.show()
```
