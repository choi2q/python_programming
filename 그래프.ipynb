```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import requests
import pandas as pd

def download_data(url):
    response = requests.get(url)
    data = response.json()
    df = pd.DataFrame(data["response"]["docs"])
    df = df["doc"].apply(pd.Series)
    return df

# 2017년 5월 데이터 다운로드
url_2017 = 'http://data4library.kr/api/loanItemSrch?libCode=111344&format=json&startDt=2017-05-01&endDt=2017-05-31&dtl_region=11040&authKey=c01ec15e4574f74ee45cba2601bad15b82971e606e3b0740977ee4b363ce2fe2'
df_2017 = download_data(url_2017)

# 2023년 5월 데이터 다운로드
url_2023 = 'http://data4library.kr/api/loanItemSrch?libCode=111344&format=json&startDt=2023-05-01&endDt=2023-05-31&dtl_region=11040&authKey=c01ec15e4574f74ee45cba2601bad15b82971e606e3b0740977ee4b363ce2fe2'
df_2023 = download_data(url_2023)

# 'loan_count' 열을 숫자 형식으로 변환
df_2017['loan_count'] = pd.to_numeric(df_2017['loan_count'])
df_2023['loan_count'] = pd.to_numeric(df_2023['loan_count'])

# 'no' 열을 기준으로 데이터프레임 병합
merged_df = pd.merge(df_2017, df_2023, on='no', suffixes=('_2017', '_2023'))

# 'loan_count_2017' 열을 기준으로 데이터프레임 내림차순 정렬 후 상위 50권 선택
top_50_books_2017 = merged_df.sort_values('loan_count_2017', ascending=False).head(50)

# 2017년과 2023년 상위 50권의 대출 횟수 비교
top_50_books_2017['loan_count_2023'] = merged_df.loc[top_50_books_2017.index, 'loan_count_2023']

# 2017년과 2023년의 대출 횟수 차이 계산
top_50_books_2017['count_change'] = top_50_books_2017['loan_count_2023'] - top_50_books_2017['loan_count_2017']

# 대출 횟수가 증가한 도서와 감소한 도서로 분리
increased_books = top_50_books_2017[top_50_books_2017['count_change'] > 0]
decreased_books = top_50_books_2017[top_50_books_2017['count_change'] < 0]

# 나눔고딕 폰트 다운로드 및 설정
font_url = 'https://cdn.jsdelivr.net/gh/google/fonts/ofl/nanumgothic/NanumGothic-Regular.ttf'
font_path = '/path/to/NanumGothic-Regular.ttf'  # 다운로드한 폰트 파일의 저장 경로로 변경해주세요

response = requests.get(font_url)
with open(font_path, 'wb') as f:
    f.write(response.content)

# 글꼴 설정을 다운로드한 글꼴 파일로 변경
fontprop = fm.FontProperties(fname=font_path)

# 각 도서의 대출 횟수 차이를 막대 그래프로 표시
plt.figure(figsize=(12, 6))
plt.bar(increased_books['bookname_2017'], increased_books['count_change'], color='green', label='증가')
plt.bar(decreased_books['bookname_2017'], decreased_books['count_change'], color='red', label='감소')
plt.xticks(rotation=90, fontproperties=fontprop)
plt.xlabel('도서', fontproperties=fontprop)
plt.ylabel('대출 횟수 변화', fontproperties=fontprop)
plt.title('2017년과 2023년 5월의 상위 50권 도서 대출 횟수 비교', fontproperties=fontprop)
plt.legend()
plt.show()
위 코드에서 '/path/to/NanumGothic-Regular.ttf'를 다운로드한 폰트 파일의 저장 경로로 변경해주세요. 이렇게 하면 나눔고딕 폰트를 사용하여 그래프를 생성할 수 있습니다.





응답 재생성


Alt+W
```

```python
import matplotlib.pyplot as plt
import requests
import pandas as pd

plt.rc('font', family='NanumBarunGothic')

def download_data(url):
    response = requests.get(url)
    data = response.json()
    df = pd.DataFrame(data["response"]["docs"])
    df = df["doc"].apply(pd.Series)
    return df

# 2017년 5월 데이터 다운로드
url_2017 = 'http://data4library.kr/api/loanItemSrch?libCode=111344&format=json&startDt=2017-05-01&endDt=2017-05-31&dtl_region=11040&authKey=c01ec15e4574f74ee45cba2601bad15b82971e606e3b0740977ee4b363ce2fe2'
df_2017 = download_data(url_2017)

# 2023년 5월 데이터 다운로드
url_2023 = 'http://data4library.kr/api/loanItemSrch?libCode=111344&format=json&startDt=2023-05-01&endDt=2023-05-31&dtl_region=11040&authKey=c01ec15e4574f74ee45cba2601bad15b82971e606e3b0740977ee4b363ce2fe2'
df_2023 = download_data(url_2023)

# 'loan_count' 열을 숫자 형식으로 변환
df_2017['loan_count'] = pd.to_numeric(df_2017['loan_count'])
df_2023['loan_count'] = pd.to_numeric(df_2023['loan_count'])

# 'no' 열을 기준으로 데이터프레임 병합
merged_df = pd.merge(df_2017, df_2023, on='no', suffixes=('_2017', '_2023'))

# 'loan_count_2017' 열을 기준으로 데이터프레임 내림차순 정렬 후 상위 50권 선택
top_50_books_2017 = merged_df.sort_values('loan_count_2017', ascending=False).head(50)

# 2017년과 2023년 상위 50권의 대출 횟수 비교
top_50_books_2017['loan_count_2023'] = merged_df.loc[top_50_books_2017.index, 'loan_count_2023']

# 2017년과 2023년의 대출 횟수 차이 계산
top_50_books_2017['count_change'] = top_50_books_2017['loan_count_2023'] - top_50_books_2017['loan_count_2017']

# 대출 횟수가 증가한 도서와 감소한 도서로 분리
increased_books = top_50_books_2017[top_50_books_2017['count_change'] > 0]
decreased_books = top_50_books_2017[top_50_books_2017['count_change'] < 0]

# 각 도서의 대출 횟수 차이를 막대 그래프로 표시
plt.figure(figsize=(12, 6))
plt.bar(increased_books['bookname_2017'], increased_books['count_change'], color='green', label='증가')
plt.bar(decreased_books['bookname_2017'], decreased_books['count_change'], color='red', label='감소')
plt.xticks(rotation=90)
plt.xlabel('도서')
plt.ylabel('대출 횟수 변화')
plt.title('2017년과 2023년 5월의 상위 50권 도서 대출 횟수 비교')
plt.legend()
plt.rcParams['axes.unicode_minus'] = False  # 음수 기호(-) 표시 설정
plt.savefig('book_loan_comparison.png', dpi=300, bbox_inches='tight', pad_inches=0.5, transparent=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import requests
import pandas as pd

plt.rc('font', family='NanumBarunGothic')

def download_data(url):
    response = requests.get(url)
    data = response.json()
    df = pd.DataFrame(data["response"]["docs"])
    df = df["doc"].apply(pd.Series)
    return df

# 2017년 5월 데이터 다운로드
url_2017 = 'http://data4library.kr/api/loanItemSrch?libCode=111344&format=json&startDt=2017-05-01&endDt=2017-05-31&dtl_region=11040&authKey=c01ec15e4574f74ee45cba2601bad15b82971e606e3b0740977ee4b363ce2fe2'
df_2017 = download_data(url_2017)

# 2023년 5월 데이터 다운로드
url_2023 = 'http://data4library.kr/api/loanItemSrch?libCode=111344&format=json&startDt=2023-05-01&endDt=2023-05-31&dtl_region=11040&authKey=c01ec15e4574f74ee45cba2601bad15b82971e606e3b0740977ee4b363ce2fe2'
df_2023 = download_data(url_2023)

# 'loan_count' 열을 숫자 형식으로 변환
df_2017['loan_count'] = pd.to_numeric(df_2017['loan_count'])
df_2023['loan_count'] = pd.to_numeric(df_2023['loan_count'])

# 출판사별 대출 횟수 계산
loan_count_2017_by_publisher = df_2017.groupby('publisher')['loan_count'].sum()
loan_count_2023_by_publisher = df_2023.groupby('publisher')['loan_count'].sum()

# 출판사별 대출 횟수 차이 계산
loan_count_change_by_publisher = loan_count_2023_by_publisher - loan_count_2017_by_publisher

# 대출 횟수가 증가한 출판사와 감소한 출판사로 분리
increased_publishers = loan_count_change_by_publisher[loan_count_change_by_publisher > 0]
decreased_publishers = loan_count_change_by_publisher[loan_count_change_by_publisher < 0]

# 출판사별 대출 횟수 차이를 막대 그래프로 표시
plt.figure(figsize=(12, 6))
plt.bar(increased_publishers.index, increased_publishers, color='green', label='증가')
plt.bar(decreased_publishers.index, decreased_publishers, color='red', label='감소')
plt.xticks(rotation=90)
plt.xlabel('출판사')
plt.ylabel('대출 횟수 변화')
plt.title('2017년과 2023년 5월의 출판사별 도서 대출 횟수 비교')
plt.legend()
plt.rcParams['axes.unicode_minus'] = False  # 음수 기호(-) 표시 설정
plt.savefig('book_loan_comparison.png', dpi=300, bbox_inches='tight', pad_inches=0.5, transparent=True)
plt.show()
```

```python
# 출판사별 대출 횟수 계산
loan_count_2017_by_publisher = df_2017.groupby('publisher').sum()['loan_count']
loan_count_2023_by_publisher = df_2023.groupby('publisher').sum()['loan_count']

# 출판사별 대출 횟수 차이 계산
loan_count_change_by_publisher = loan_count_2023_by_publisher - loan_count_2017_by_publisher

# 데이터프레임 생성
publisher_df = pd.DataFrame({
    'publisher': loan_count_change_by_publisher.index,
    'loan_count_change': loan_count_change_by_publisher.values
})

# 대출 횟수가 증가한 출판사와 감소한 출판사로 분리
increased_publishers = publisher_df[publisher_df['loan_count_change'] > 0]
decreased_publishers = publisher_df[publisher_df['loan_count_change'] < 0]

# 출판사별 대출 횟수 변화를 산점도로 표시
plt.figure(figsize=(8, 8))
plt.scatter(increased_publishers['loan_count_change'], increased_publishers['publisher'], color='green', label='증가')
plt.scatter(decreased_publishers['loan_count_change'], decreased_publishers['publisher'], color='red', label='감소')
plt.xlabel('대출 횟수 변화')
plt.ylabel('출판사')
plt.title('2017년과 2023년 출판사별 도서 대출 횟수 변화')
plt.legend()
plt.grid(True)
plt.savefig('publisher_loan_change.png', dpi=300, bbox_inches='tight', pad_inches=0.5, transparent=True)
plt.show()
```
